## Selecionando com múltiplas condições

```python
# Selecionando operador loc com multiplas condições
data[['Idade','Sexo','Tarifa','Classe','Sobreviveu']].loc[(data['Idade']>=3)&(data['Idade']<=10)\
    &(data['Sobreviveu']==1)]
```

## map e replace

```python
# É possível selecionar as colunas só com um ponto- Data series
data.Sexo
```

```python
# Outra forma de Data series
data['Sexo']
```

```python
# Data frame
data[['Sexo']]
```

```python
# Através do map é possível substituir com dicionário nomes de colunas(somente visualização)
data['Sexo'].map({'male': 'homem','female':'mulher'}).head(16)
```

```python
# Outra maneira de substituir nomes
data['Sexo'].replace({'male': 'homem','female':'mulher'}).head()
```

```python
# Terceira maneira de substituir nomes
data['Sexo'].replace(['male','female'],['homem','mulher'], inplace=True)
```

```python
data['Sexo']
```
